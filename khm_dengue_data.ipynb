```python
import numpy as np
import pandas as pd

csv_file_path = 'data/KHM/cambodia_dengue/DengueNet_Cases_Cambodia_1998_2010.csv'
```

```python
def convert_date(raw_date):
    import re
    result = None
    date_regex = r'(\d{4})\s+month:\s+(\d{1,2})'
    matches = re.search(date_regex, raw_date)
    
    if matches:
        year = matches.group(1)
        month = matches.group(2)
        result = '{0}-{1}'.format(year, month)
    
    return result
```

```python
df = pd.read_csv(csv_file_path, header=None, sep=",")
arr = df.as_matrix().T

dates_raw = arr[3:, 0]
dates = [convert_date(x) for x in dates_raw]
dates = np.insert(dates, 0, ['DATE'])
arr = arr[2:,:]
arr = arr[:,1:]

arr = np.c_[dates, arr]
column_names = arr[0,:]
arr = arr[1:,:]
df = pd.DataFrame(arr, columns=column_names)
df = df.apply(lambda x: pd.to_numeric(x, errors='ignore'))
df['DATE'] = pd.to_datetime(df['DATE'], format='%Y-%m')
df.head()

```

```python
df.to_csv('khm_denguenet_data_cleaned.csv')
```
