```python
import pandas as pd
import csv

csv_file_path = 'old_tweets.csv'
```

```python
df = pd.read_csv(csv_file_path, header=0, sep=',', parse_dates=[1], quoting=csv.QUOTE_ALL, encoding='utf-8')
df.set_index(['id'])
df.head()
```

```python
df.describe()
```

```python
#filter the tweets that are not in South or Central America
df = df[df['lat'] < 23]
df = df[df['lat'] > -56]
df = df[df['long'] < -34]
df = df[df['long'] > -107]

print(df['id'].count())
```

```python
df.to_csv('tweets_south_america.csv')
```

```python
df2 = pd.read_csv('cities.csv', names=['place','approx_lat', 'approx_long'], header=0, sep=',', quoting=csv.QUOTE_NONNUMERIC, encoding='utf-8')
df2.head()
```

```python
df2[df2['place'] == 'Brasília, Brazil']['approx_lat'].values[0]
```

```python
df['approx_lat'] = pd.Series(df['geo'])
df['approx_long'] = pd.Series(df['geo'])

df['approx_lat'] = df['approx_lat'].map(lambda geo : df2[df2['place'] == geo]['approx_lat'].values[0])
df['approx_long'] = df['approx_long'].map(lambda geo : df2[df2['place'] == geo]['approx_long'].values[0])

df.head()
```

```python
df.to_csv('tweets_south_america_approx.csv')
```
